```python
from xml.dom.minidom import parse
import xml.dom.minidom
import string

DOMTree = xml.dom.minidom.parse("../JonoBot/.venv/BlogExport.xml")
collection = DOMTree.documentElement
```

```python
# Get everything I can out of the posts and pages on JonathanCavell.com.
items = collection.getElementsByTagName("item")
posts = []

for item in items:
    if (item.getElementsByTagName('wp:post_type')[0].childNodes[0].data == "page"):
        if (len(item.getElementsByTagName('content:encoded')[0].childNodes) > 0):
            text = ''
            text = text + item.getElementsByTagName('content:encoded')[0].childNodes[0].data
            text = text.replace('<!-- wp:paragraph -->', '')
            text = text.replace('\n', '')
            text = text.replace('<p>', '')
            text = text.replace('<em>', '')
            text = text.replace('</p>', '')
            text = text.replace('<!-- /wp:paragraph -->', '')
            text = text.replace('<h1>', '')
            text = text.replace('</h1>', '')
            text = text.replace('\t', '')
            if text.find("wp:group") == -1 and text.find("wp:image") == -1 and text.find("<style>") == -1 and text.find("hello-world") == -1 and text.find("<h1>Blog</h1>") == -1:
                posts.append(text)
        
    if (item.getElementsByTagName('wp:post_type')[0].childNodes[0].data == "post"):
        if (len(item.getElementsByTagName('content:encoded')[0].childNodes) > 0):
            text = ''
            text = text + item.getElementsByTagName('content:encoded')[0].childNodes[0].data
            text = text.replace('<!-- wp:paragraph -->', '')
            text = text.replace('\n', '')
            text = text.replace('<p>', '')
            text = text.replace('<em>', '')
            text = text.replace('</p>', '')
            text = text.replace('<!-- /wp:paragraph -->', '')
            text = text.replace('<h1>', '')
            text = text.replace('</h1>', '')
            text = text.replace('\t', '')
            if text.find("wp:group") == -1 and text.find("wp:image") == -1:
                posts.append(text)

for post in posts:
  display(post)
```

```python
import json

training_data = []

for post in posts:
  line = {"input": post}
  training_data.append(json.dumps(line))
  training_data.append("\n")

with open("training_data_from_blog.jsonl", "w") as f:
  f.writelines(training_data)


```
